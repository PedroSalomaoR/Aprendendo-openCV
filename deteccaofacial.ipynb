```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

haarcascade_path = cv.data.haarcascades + "haarcascade_frontalface_default.xml"
print("Caminho do arquivo:", haarcascade_path)

face_classifier = cv.CascadeClassifier(haarcascade_path)

image = cv.imread('jogo-corinthians-hoje-7-scaled-aspect-ratio-512-320.webp')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

faces = face_classifier.detectMultiScale(gray, 1.3, 5)

for (x, y, w, h) in faces:
    cv.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

cv2_imshow(image)

```
