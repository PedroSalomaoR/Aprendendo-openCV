```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Lenna.png')

b, g, r = cv2.split(img)
zeros = np.zeros_like(b)

azul = cv2.merge([b, zeros, zeros])
verde = cv2.merge([zeros, g, zeros])
vermelho = cv2.merge([zeros, zeros, r])

plt.figure(figsize=(10,4))
plt.subplot(1,3,1), plt.imshow(azul)
plt.title('Azul'), plt.axis('off')
plt.subplot(1,3,2), plt.imshow(verde)
plt.title('Verde'), plt.axis('off')
plt.subplot(1,3,3), plt.imshow(vermelho)
plt.title('Vermelho'), plt.axis('off')
plt.show()
```
